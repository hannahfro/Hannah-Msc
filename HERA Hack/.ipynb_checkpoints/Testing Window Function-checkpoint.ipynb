# Define a theory spectrum

```python
def ps(x):
    mu, sigma = 0.05,0.007
    return  np.exp( - (x - mu)**2 / (2 * sigma**2))
```

# Compute and Plot Window Function 

```python
pk, kk = window.compute_pspec_estimate(ps)
```

```python
fig, ax = plt.subplots(2, 1, figsize = (15,6),sharex='col', sharey='row', gridspec_kw={'hspace': 0, 'wspace': 0})

ax[0].plot(kk,ps(kk), label = 'theory binned spectrum')
ax[0].plot(kk,pk, label = 'window estimated spectrum ')
ax[0].set_xlabel('k')
ax[0].set_ylabel('P(k)')
ax[0].legend()
ax[0].set_title('Power Spectra and Window Functions')
for i in range(window.nbins): 
    ax[1].plot(kk,window.window_binned[i],c = 'k')
ax[1].set_xlabel('k')
```
