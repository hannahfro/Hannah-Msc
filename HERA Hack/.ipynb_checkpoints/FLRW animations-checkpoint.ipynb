```python

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np



theta_s = np.linspace(0, 2*np.pi, 100)   #setting range for theta
phi_s = np.linspace(0, np.pi, 100)       #setting range for phi
phi_s, theta_s = np.meshgrid(phi_s, theta_s)   #setting the grid for phi and theta

#Setting the cartesian coordinates of the unit sphere
#Converting phi, theta, z to cartesian coordinates
x_s = np.sin(phi_s)*np.cos(theta_s)
y_s = np.sin(phi_s)*np.sin(theta_s)
z_s = np.cos(phi_s)

theta = np.linspace(0, 2*np.pi, 15)   #setting range for theta
phi = np.linspace(0, np.pi, 15)       #setting range for phi
phi, theta = np.meshgrid(phi, theta)   #setting the grid for phi and theta

#Setting the cartesian coordinates of the unit sphere
#Converting phi, theta, z to cartesian coordinates
x = np.sin(phi)*np.cos(theta)
y = np.sin(phi)*np.sin(theta)
z = np.cos(phi)


fig = plt.figure(figsize=[20,20])
ax = fig.gca(projection='3d')
ax.plot_surface(x_s, y_s, z_s, alpha=0.5, color = 'k')
ax.plot_wireframe(x, y, z, alpha=0.5, color = 'k')

ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])

# make the panes transparent
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# make the grid lines transparent
ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)

for tic in ax.xaxis.get_major_ticks():
    tic.tick1On = tic.tick2On = False
    tic.label1On = tic.label2On = False

for angle in np.linspace(0, 360, 360):

    ax.view_init(-140, angle)
    plt.draw()

    #Turn off the axis planes so only the sphere is visible


    plt.savefig('sphere_%04d.png' % angle)          #saves a .png file of my figure
```

```python
import matplotlib.pyplot as plt
import numpy as np
from itertools import product, combinations

fig = plt.figure(figsize = [30,30])

scale_factor = np.linspace(1,3,3)

points = np.asarray([np.random.uniform(-1,1,20),np.random.uniform(-1,1,20),np.random.uniform(-1,1,20)])

# draw cube



for i in range(len(scale_factor)):
    scale_vec = np.ones(len(points[0]))*scale_factor[i]
    r = [-1*scale_factor[i], 1*scale_factor[i]]
    ax = fig.gca(projection='3d')
    ax.set_xlim(-3,3)
    ax.set_ylim(-3,3)
    ax.set_zlim(-3,3)
    ax.scatter3D(scale_vec*points[0],scale_vec*points[1],scale_vec*points[2],alpha = 1, color = 'k')

    for s, e in combinations(np.array(list(product(r, r, r))), 2):
        if np.sum(np.abs(s-e)) == r[1]-r[0]:
            ax.plot3D(*zip(s, e), color="black")
            
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_zticklabels([])


    # make the panes transparent
    ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    # make the grid lines transparent
    ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
    ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
    ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)

    for tic in ax.xaxis.get_major_ticks():
        tic.tick1On = tic.tick2On = False
        tic.label1On = tic.label2On = False


    plt.savefig('lightexpanding_%04d.png' % i)          #saves a .png file of my figure
    
plt.show()
```
